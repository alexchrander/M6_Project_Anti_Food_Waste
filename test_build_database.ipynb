```python
# Sort and get last snapshot per offer
history['fetched_at'] = pd.to_datetime(history['fetched_at'])
history['offer_end_time'] = pd.to_datetime(history['offer_end_time'])

last_snapshot = (
    history.sort_values('fetched_at')
    .groupby('unique_id')
    .last()
    .reset_index()
)

# How early did each offer disappear relative to its end_time?
last_snapshot['minutes_before_end'] = (
    last_snapshot['offer_end_time'] - last_snapshot['fetched_at']
).dt.total_seconds() / 60

# Distribution of this
last_snapshot['minutes_before_end'].describe()
```

```python
# How many snapshots does each offer have?
snapshot_counts = history.groupby('unique_id')['fetched_at'].count().rename('n_snapshots')

# How long was each offer observed (first to last snapshot)?
offer_span = history.groupby('unique_id')['fetched_at'].agg(['min', 'max'])
offer_span['observed_hours'] = (offer_span['max'] - offer_span['min']).dt.total_seconds() / 3600

# Stock trajectory — did stock drop before disappearing?
first_stock = history.sort_values('fetched_at').groupby('unique_id')['offer_stock'].first().rename('first_stock')
last_stock = history.sort_values('fetched_at').groupby('unique_id')['offer_stock'].last().rename('last_stock')

offer_summary = pd.concat([snapshot_counts, offer_span, first_stock, last_stock], axis=1)
offer_summary['stock_dropped'] = offer_summary['first_stock'] - offer_summary['last_stock']

print(offer_summary[['n_snapshots', 'observed_hours', 'first_stock', 'last_stock', 'stock_dropped']].describe())
```

```python
print(offer_summary['last_stock'].value_counts().sort_index().head(20))

# And what share of offers had zero stock drop?
no_drop = (offer_summary['stock_dropped'] == 0).mean()
print(f"\nOffers with no stock drop: {no_drop:.1%}")

# Cross: last_stock vs whether any drop occurred
print(offer_summary.groupby('last_stock')['stock_dropped'].agg(['count', 'mean']).head(15))
```

```python
# For offers that DID have stock drop, what did the trajectory look like?
# Specifically - of offers ending at last_stock=1, how many had a drop vs didn't?
ended_at_1 = offer_summary[offer_summary['last_stock'] == 1.0]
print(ended_at_1['stock_dropped'].value_counts().sort_index().head(20))
print(f"\nEnded at 1 with some drop: {(ended_at_1['stock_dropped'] > 0).mean():.1%}")
```

```python
completed = history[~history['unique_id'].isin(current['unique_id'])]

lifecycle_stats = completed.groupby('unique_id').agg(
    first_seen=('fetched_at', 'min'),
    last_seen=('fetched_at', 'max'),
    first_stock=('offer_stock', 'first'),
    last_stock=('offer_stock', 'last'),
    n_snapshots=('fetched_at', 'count')
).reset_index()

lifecycle_stats['hours_on_clearance'] = (
    lifecycle_stats['last_seen'] - lifecycle_stats['first_seen']
).dt.total_seconds() / 3600

lifecycle_stats['sell_through_rate'] = (
    1 - (lifecycle_stats['last_stock'] / lifecycle_stats['first_stock'])
)

print(lifecycle_stats[['hours_on_clearance', 'sell_through_rate']].describe())
print(f"\nZero sell-through: {(lifecycle_stats['sell_through_rate'] == 0).mean():.1%}")
print(f"100% sell-through: {(lifecycle_stats['sell_through_rate'] == 1).mean():.1%}")
```

```python
# Then the overnight survival check
history['fetched_at'] = pd.to_datetime(history['fetched_at'])
history['date'] = history['fetched_at'].dt.date

# Get unique dates in the data
dates = sorted(history['date'].unique())
print(f"Dates in data: {dates}")

# For each consecutive day pair, find offers alive near end of day 1
# and check if they reappear at start of day 2
results = []
for i in range(len(dates) - 1):
    day1, day2 = dates[i], dates[i+1]
    
    # Offers seen in last 2 hours of day1
    day1_data = history[history['date'] == day1]
    day1_last2h = day1_data[day1_data['fetched_at'] >= day1_data['fetched_at'].max() - pd.Timedelta(hours=2)]
    alive_eod = set(day1_last2h['unique_id'].unique())
    
    # Offers seen in first 2 hours of day2
    day2_data = history[history['date'] == day2]
    day2_first2h = day2_data[day2_data['fetched_at'] <= day2_data['fetched_at'].min() + pd.Timedelta(hours=2)]
    alive_next_morning = set(day2_first2h['unique_id'].unique())
    
    survived = alive_eod & alive_next_morning
    results.append({
        'day1': day1, 'day2': day2,
        'alive_eod': len(alive_eod),
        'survived_to_morning': len(survived),
        'survival_rate': len(survived) / len(alive_eod) if alive_eod else 0
    })

print(pd.DataFrame(results))
```

### Building the dataset

```python
import pandas as pd
import sqlite3
from pathlib import Path

conn = sqlite3.connect("data/food_waste.db")
history = pd.read_sql("SELECT * FROM history", conn)
current = pd.read_sql("SELECT * FROM current", conn)
conn.close()

# Parse timestamps
history['fetched_at'] = pd.to_datetime(history['fetched_at'])
history['offer_start_time'] = pd.to_datetime(history['offer_start_time'])
history['offer_end_time'] = pd.to_datetime(history['offer_end_time'])

# Exclude active offers
active_ids = set(current['unique_id'])
completed = history[~history['unique_id'].isin(active_ids)].copy()

print(f"Total unique offers in history: {history['unique_id'].nunique()}")
print(f"Excluded (still active): {len(active_ids)}")
print(f"Completed offers: {completed['unique_id'].nunique()}")
```

```python
# Sort once upfront
completed = completed.sort_values(['unique_id', 'fetched_at'])

# First and last snapshots per offer
first = completed.groupby('unique_id').first().reset_index()
last = completed.groupby('unique_id').last().reset_index()

# Overnight gap flag — any consecutive gap > 4 hours
completed['time_diff_hours'] = (
    completed.groupby('unique_id')['fetched_at']
    .diff()
    .dt.total_seconds() / 3600
)
overnight_gap = (
    completed.groupby('unique_id')['time_diff_hours']
    .max()
    .gt(4)
    .reset_index()
    .rename(columns={'time_diff_hours': 'had_overnight_gap'})
)

# Snapshot count
n_snapshots = (
    completed.groupby('unique_id')['fetched_at']
    .count()
    .reset_index()
    .rename(columns={'fetched_at': 'n_snapshots'})
)

# Build dataset from first snapshot columns
dataset = first[[
    'unique_id', 'store_id', 'store_name', 'store_brand', 'store_city',
    'product_ean', 'product_description', 'offer_ean',
    'offer_original_price', 'offer_new_price', 'offer_discount',
    'offer_percent_discount', 'offer_stock_unit', 'offer_start_time',
    'offer_end_time', 'store_customer_flow',
    'product_category_da', 'product_category_en',
]].copy()

dataset['initial_stock'] = first['offer_stock']
dataset['final_stock'] = last['offer_stock']
dataset['first_seen'] = first['fetched_at']
dataset['last_seen'] = last['fetched_at']

# Labels
dataset['sell_through_rate'] = (
    1 - (dataset['final_stock'] / dataset['initial_stock'])
).clip(0, 1)

dataset['hours_on_clearance'] = (
    dataset['last_seen'] - pd.to_datetime(dataset['offer_start_time'])
).dt.total_seconds() / 3600

# Merge metadata
dataset = dataset.merge(n_snapshots, on='unique_id')
dataset = dataset.merge(overnight_gap, on='unique_id')

print(dataset.shape)
print(dataset[['sell_through_rate', 'hours_on_clearance', 'n_snapshots', 'had_overnight_gap']].describe())
```

```python
# Use product_ean instead of offer_ean
offer_timeline = dataset[['unique_id', 'store_id', 'product_ean', 'first_seen', 'last_seen', 'sell_through_rate']].copy()

merged = offer_timeline.merge(
    offer_timeline,
    on=['store_id', 'product_ean'],
    suffixes=('_prev', '_next')
)

relabelled = merged[
    (merged['unique_id_prev'] != merged['unique_id_next']) &
    (merged['first_seen_next'] > merged['last_seen_prev']) &
    (merged['first_seen_next'] - merged['last_seen_prev'] < pd.Timedelta(hours=2))
]

print(f"Potential relabelling events: {len(relabelled)}")
print(f"Unique offers potentially relabelled: {relabelled['unique_id_prev'].nunique()}")
print()
if len(relabelled) > 0:
    print(relabelled[['store_id', 'product_ean', 'last_seen_prev', 'first_seen_next', 'sell_through_rate_prev']].head(10))
```
